```python
url_base='http://sistemas.tce.pe.gov.br/DadosAbertos/'
```

## criar url

```python
def urlset(metodo, campo1=None, campo2=None):
    if campo1 is None:
        campo1=campo2=""
    elif campo2 is None:
        campo2=""
    
    url = url_base + metodo + "!json?" + campo1 + campo2
    return url
```

## municipios

```python
url=urlset(munic)
with urlopen(url) as url:
    parsed = json.loads(url.read().decode('latin-1'))
data = json.dumps(parsed, indent=4, sort_keys=True, ensure_ascii=False)
print(data)
```
